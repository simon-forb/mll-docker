# MLL - Machine Learning Language

## Generate simple expression and evaluate.

```java
import mll.*;

// Generate example DAG
DAG dag = new DAG();
Op x = dag.x();
Op a = dag.lit(3.f);
Op add = x.add(a);

// Evaluate
System.out.println(add.eval(2.f));
```

## Save as PNG and display inline.

```java
// Save DOT as PNG
MLLRunner.saveDot(add, "test");

// View DOT in Notebook
MLLRunner.viewDot(add);
```

## Evaluate for Multiple Values and Plot

```java
// Generate sample plot data
float[] xValues = new float[5];
float[] yValues = new float[5];
for (int i = 0; i < 5; i++) {
  xValues[i] = i;
  yValues[i] = add.eval((float) i);
}

// Save function plot to disk
MLLRunner.saveFunctionPlot(xValues, yValues, "plot");

// View plot in Jupyter-Notebook
MLLRunner.viewFunctionPlot(xValues, yValues, "plot");
```

## Generate LLVM

```java
// Save LLVM to disk
add.llvmToFile("program");

// Show LLVM in Jupyter-Notebook
MLLRunner.viewLLVM(add);
```

## Optimize LLVM

```java
// Save optimized LLVM to disk
MLLRunner.saveOpt("program", 1, true);

// Show optimized LLVM in Jupyter-Notebook
MLLRunner.viewOpt(add, 3, true);
```

## Run LLVM

```java
MLLRunner.runLLVM("program", true);
// Program fails due to missing main method
```
